```python
!pip install pandas openpyxl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "All_games_vf1.xlsx"
df = pd.read_excel(file_path)
df
```

```python
sns.heatmap(df[['Days of Operation','revenue','price', 'copiesSold', 'reviewScore']].corr(),
            annot=True,
            fmt=".2f",
            cmap='coolwarm')

plt.title('Correlation Matrix')
```

```python
df_ind = df[df['Genre'] == 'Ind']
sns.heatmap(df_ind[['Days of Operation','revenue','price', 'copiesSold', 'reviewScore']].corr(),
            annot=True,
            fmt=".2f",
            cmap='coolwarm')

plt.title('Correlation Matrix Indie Game')
```

```python
df_act = df[df['Genre'] == 'Act']
sns.heatmap(df_act[['Days of Operation','revenue','price', 'copiesSold', 'reviewScore']].corr(),
            annot=True,
            fmt=".2f",
            cmap='coolwarm')

plt.title('Correlation Matrix Action Game')
```

```python
df_cas = df[df['Genre'] == 'Cas']
sns.heatmap(df_cas[['Days of Operation','revenue','price', 'copiesSold', 'reviewScore']].corr(),
            annot=True,
            fmt=".2f",
            cmap='coolwarm')

plt.title('Correlation Matrix Casual Game')
```
